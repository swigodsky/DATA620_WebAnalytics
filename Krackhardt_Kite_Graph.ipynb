```python
#import libraries
import networkx as net
import networkx.generators.small
import matplotlib.pyplot as plot
```

```python
#generage krackrardt kite graph
g = networkx.generators.small.krackhardt_kite_graph(create_using=None)

#number of edges
g.number_of_edges()
```

```python
#number of nodes
g.number_of_nodes()
```

```python
#change node names from numbers to names
mapping={0:'Andre',1:'Beverly',2:'Carol',3:'Diane',4:'Ed',5:'Fernando',6:'Garth',7:'Heather',8:'Ike',9:'Jane'}
h=net.relabel_nodes(g,mapping)

#draw kite graph
net.draw(h, with_labels=True)
```
